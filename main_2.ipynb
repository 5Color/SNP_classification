```python
import pandas as pd
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv("./train.csv")
test = pd.read_csv("./test.csv")
train = pd.DataFrame(train)
test = pd.DataFrame(test)
```

```python
train
```

```python
train_x = train.drop(['id', 'class'], axis=1)

le = LabelEncoder()
for col in train_x.columns:
    if train_x[col].dtype == 'object':  # 문자열 타입만 변환
        train_x[col] = le.fit_transform(train_x[col])
```

```python
test = test.drop(['id'], axis=1)
for col in test.columns:
    if test[col].dtype == 'object':  # 문자열 타입만 변환
        test[col] = le.fit_transform(test[col])
```

```python
train_y = train['class']
```

```python
le = LabelEncoder()
train_y = le.fit_transform(train_y)
```

```python
train_input, test_input, train_target, test_target = train_test_split(train_x, train_y, test_size=0.2, stratify=train_y)

train_input
```

```python
train_target
```

```python
lgbm = LGBMClassifier()
lgbm.fit(train_input, train_target)
```

```python
preds = lgbm.predict(test)
preds
```

```python
submit2 = pd.read_csv('./sample_submission.csv')
submit2['class'] = le.inverse_transform(preds)
```

```python
submit2.to_csv("./submit2.csv", index=False)
```
